```python
with open('sherlock.txt', 'rt') as f:
    content = f.read()
```

```python
words = content.split()
```

```python
import re
m = re.compile(r'[a-z]+$')

words = [word.lower() for word in words if len(word)==5]
words = [word for word in words if m.match(word)]
```

```python
from collections import Counter
Counter(words).most_common()
```

```python
len(words)
```

```python
words = [word[0] for word in Counter(words).most_common()]

Counter(''.join(words)).most_common()
```

```python

words = [word[0] for word in Counter(words).most_common(50)]
words[:5]
```
